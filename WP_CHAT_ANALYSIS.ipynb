```python
import re
import pandas as pd
```

```python
f=open('wp_chat_cybg.txt','r',encoding='utf-8')
```

```python
data=f.read()
```

```python
print(data )
```

```python
pattern = r'\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{1,2}\s*(?:AM|PM)\s-\s'
```

```python
massage=re.split(pattern,data)[1:]
massage
```

```python
dates=re.findall(pattern, data)
dates
```

```python
Dates=[]
for message in dates:
    cleaned_message = message.replace("\u202f", " ")
    Dates.append(cleaned_message)
```

```python
Dates
```

```python
df=pd.DataFrame({'user_massage':massage,'massage_date':Dates})
df['massage_date']=pd.to_datetime(df['massage_date'],format='%m/%d/%y, %I:%M %p - ')
df.rename(columns={'massage_date':'date'},inplace=True)
df.head()
```

```python
users=[]
messages=[]
for message in df['user_massage']:
  entry=re.split('([\w\W]+?):\s',message)
  if entry[1:]:
    users.append(entry[1])
    messages.append(entry[2])
  else:
    users.append('group_notification')
    messages.append(entry[0])
df['users']=users
df['massage']=messages
df.drop(columns=['user_massage'],inplace=True)
```
